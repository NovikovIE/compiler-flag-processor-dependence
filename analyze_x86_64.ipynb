```python
import pandas as pd
import numpy as np
```

```python
df_cbench = pd.read_csv("cbench_results_x86_64.csv").reset_index()
df_mibench = pd.read_csv("mibench_results_x86_64.csv").reset_index()

df = pd.concat([df_cbench, df_mibench], axis=0)

time_column = 'median_exec_time_sec'
df[time_column] = pd.to_numeric(df[time_column])
```

```python
benchmarks = df['benchmark'].unique()
benchmarks
```

```python
for benchmark_name in benchmarks:
    print(f"\n--- Бенчмарк: {benchmark_name} ---")

    # Фильтруем данные только для текущего бенчмарка
    bench_df = df[df['benchmark'] == benchmark_name]

    # Находим строку с лучшим (минимальным) временем выполнения
    best_run = bench_df.loc[bench_df[time_column].idxmin()]
    best_time = best_run[time_column]
    best_flags = best_run.drop(['benchmark', time_column])
    
    best_relative_speedup = best_run['speedup_vs_O3']

    mean_time = bench_df[time_column].mean()
    std_dev = bench_df[time_column].std()

    print("  Лучший результат:")
    print(f"    Время выполнения: {best_time:.6f} сек")
    print(f"    Ускорение: {best_relative_speedup:.4f}x")
    print("    Флаги:")
    print("      ", end="")
    for flag_name, flag_value in best_flags.items():
        print(f"{flag_name}: {flag_value}", end=' ') 
        
    print("\n  Статистика по всем комбинациям:")
    print(f"    Среднее время:    {mean_time:.6f} сек")
    print(f"    Станд. отклонение: {std_dev:.6f} сек")
```
